## Making the abstract in python

```python
from rdkit import Chem, rdBase
from rdkit.Chem import Draw, Descriptors, PandasTools, AllChem
from rdkit.Chem.Draw import IPythonConsole
#Letting it be BnW
IPythonConsole.drawOptions.useBWAtomPalette()
```

```python
paroxetine = "C1CNCC(C1C2=CC=C(C=C2)F)COC3=CC4=C(C=C3)OCO4"
paroxetine_mol = Chem.MolFromSmiles(paroxetine)
paroxetine_mol
```

```python
print(f'Molecular Weight: {Descriptors.ExactMolWt(paroxetine_mol)}')
print(f'NumHBond: {Descriptors.NumHDonors(paroxetine_mol)}')
print(f'NumHAcceptor: {Descriptors.NumHAcceptors(paroxetine_mol)}')
print(f'LogP: {Descriptors.MolLogP(paroxetine_mol)}')
```
