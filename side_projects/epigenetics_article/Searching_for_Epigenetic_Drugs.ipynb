We already have 7 approved epigenetic drugs Azacitidine, Decitabine, Vorinostat, Belinostat, Panobinostat, Romidepsin, Tazemetostat

In this notebook, I want to get the molecular targest of these drugs, and check if there are any other FDA approved drugs that share the same target

```python
import pandas as pd
```

```python
# importing the dataset with molecules IDs, and molecular targets

drugs = pd.read_csv("/content/uniprot.csv")
drugs
```

```python
# importing the dataset with only FDA approved molecules

fda_approved = pd.read_excel("/content/RAW_fda_approved_1997_2021(1997_2021).xlsx")
fda_approved.head()
```

```python
# getting only the epigenetic drugs information

epigenetics = ['Azacitidine', 'Decitabine', 'Vorinostat', 'Belinostat', 
               'Panobinostat', 'Romidepsin', 'Tazemetostat']

epigenetics_drugs = drugs[drugs.Name.isin(epigenetics)]
epigenetics_drugs.head()
```

```python
# sanity check

print(epigenetics_drugs.shape)
epigenetics_drugs.Name.unique()
```

```python
#sanity check - checking if all epigenetic drugs are in the approved dataset

id_epi = epigenetics_drugs['DrugBank ID'].unique()

fda_approved[fda_approved['DrugBank ID'].isin(id_epi)]
```

```python
# getting the targets for the epigenetic drugs

targets = epigenetics_drugs['UniProt ID'].unique()
targets
```

```python
# getting all the drugs that share the same epigenetic target 

drugs_same_target = drugs[drugs['UniProt ID'].isin(targets)]
drugs_same_target.Name.unique()
```

```python
# checking which of those drugs are FDA approved

ids = drugs_same_target['DrugBank ID'].unique()

fda_approved[fda_approved['DrugBank ID'].isin(ids)]
```

We can see here that the drugs Fingolimod and Arsenic Trioxide were added in this dataset, meaning it shares a epigenetic target, and wasn't in my original epigenetic drugs list.
